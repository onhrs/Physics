```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(-3, 3, 0.25)
y = np.arange(-3, 3, 0.25)
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x_c =  np.outer(np.cos(u), np.sin(v))
y_c = np.outer(np.sin(u), np.sin(v))
z_c = np.outer(np.ones(np.size(u)), np.cos(v))

x_e =  np.outer(0.5*np.cos(u), 0.5*np.sin(v))
y_e = np.outer(0.5*np.sin(u), 0.5*np.sin(v))
z_e = np.outer(0.5*np.ones(np.size(u)),0.5* np.cos(v))


X, Y = np.meshgrid(x, y)
coef=0
Z =(coef*X**2+coef*Y**2)-18*coef

ax.plot_wireframe(X,Y,Z)
ax.plot_surface(x_c+1, y_c-2, z_c+2-10*coef, color='b')
ax.plot_surface(x_e+3.5, y_e-2, z_e+1.5, color='r')
ax.set_xlim3d([-4.0, 4.0])
ax.set_ylim3d([-4.0, 4.0])
ax.set_zlim3d([-4.0, 4.0])
ax.tick_params(labelbottom="off",bottom="off") # x
ax.tick_params(labelleft="off",left="off") # y
```
